```python
df_train2016 = pd.read_csv('../train_2016_v2.csv/train_2016_v2.csv', low_memory=False)
#Merging training data provided with the original data on the basis of parcelid
df_merge = pd.merge(df_train2016,df_prop,on="parcelid",how="left")
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
x=df_merge["logerror"]

#No data cleaning to be done in TASK 3. Only replaced NaN values with the mean. Cleaning is done in 
df_merge.bedroomcnt=df_merge.bedroomcnt.fillna(df_merge.bedroomcnt.mean())
df_merge.calculatedfinishedsquarefeet=df_merge.calculatedfinishedsquarefeet.fillna(df_merge.calculatedfinishedsquarefeet.mean())
df_merge.garagetotalsqft=df_merge.garagetotalsqft.fillna(df_merge.garagetotalsqft.mean())
df_merge.bathroomcnt=df_merge.bathroomcnt.fillna(df_merge.bathroomcnt.mean())
df_merge.taxamount=df_merge.taxamount.fillna(df_merge.taxamount.mean())

y=df_merge[['bedroomcnt', 'bathroomcnt','taxamount','calculatedfinishedsquarefeet']]
model = sm.OLS(x, y).fit()
model.summary()

#FOR TASK 3 --> Only NaN values substitution is done as part of cleaning as there is no need for the rest.
#MAJORITY OF THE CLEANING IS DONE IN TASK 4 BELOW.
```

```python
'How well/badly does it work?'
The R-squared value is 0.008 without intercept, which means that only 0.8% of the log errors can be explained by our predictor variable bedroomcnt, bathroomcnt,taxamount and calculatedfinishedsquarefeet.
Though for every predictor variabe the std error is very low. But for that metric evaluation of the variables needs to be performed.
Even after taking the z_score of the predictor variables, the results remain the same. So, normalization does not affect our model.
The r-squared value is low because even small improvements in the zillow model are hard to come by as it is already a very good model.
Thats why, I think the choosen variables work well with a r-squared value of 0.008.

'Which variable is the most important one?'
calculatedfinishedsquarefeet is the most important variable in our linear regression model.
calculatedfinishedsquarefeet stands for calculated total finished living area of the home.
This is one of the factors which influence the cost of a house. So, it is correlated to the logerror prediction.
A linear regression of logerror on calculatedfinishedsquarefeet gives a R-squared value of 0.006.
Including this variable to the model certainly increases the r-squared value and thus a better model.
```

```python
#<--------------------CLEANING PHASE-------------------->

df_clean = df_merge.copy()
#DROPPED COLUMNS --> by thresh : require that many non-NA values
df_clean = df_clean.dropna(thresh=len(df_clean)/10, axis=1)

#Fill NaN values for column calculatedfinishedsquarefeet by multiplying average area per bedroom * number of bedroom
df_clean['area_per_bedroom'] = df_clean['calculatedfinishedsquarefeet'] / df_clean['bedroomcnt']
df_clean.area_per_bedroom[np.isinf(df_clean.area_per_bedroom)] = df_clean.calculatedfinishedsquarefeet
mean_area_per_bedroom = df_clean.area_per_bedroom.mean()
df_clean.calculatedfinishedsquarefeet = df_clean.calculatedfinishedsquarefeet.fillna(mean_area_per_bedroom*df_clean.bedroomcnt)

#Removing unnecessary or redundant columns
df_clean=df_clean.drop('area_per_bedroom',axis=1)      #Self created column in the dataset
df_clean=df_clean.drop('finishedsquarefeet12',axis=1)  #Redundant with column 'calculatedfinishedsquarefeet'
df_clean=df_clean.drop('calculatedbathnbr',axis=1)     ##Redundant with column 'bathroomcnt'

#Drop columns we won't be using in our current model prediction
df_clean=df_clean.drop('latitude',axis=1)
df_clean=df_clean.drop('longitude',axis=1)

#Fill NaN values of fullbathcnt with the value of the bathroom count. Let us take that a house with no value provided for full bathroom count has all the bathrooms as full bathrooms.
df_clean.fullbathcnt = df_clean.fullbathcnt.fillna(df_clean.bathroomcnt)
df_clean.fullbathcnt.count()

#Substitute NaN values of yearbuilt by the mean of the column
df_clean.yearbuilt = df_clean.yearbuilt.fillna(np.floor(df_clean.yearbuilt.mean()))
```

```python
#Distribute the cleaned data in two parts. 80% is the training data while 20% is the testing data.
distribute = np.random.rand(len(df_clean)) < 0.8
df_train = df_clean[distribute]
df_test = df_clean[~distribute]
```

```python
#DECISION TREE MODEL
from sklearn.datasets import load_boston
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)

X = df_train.filter(['bedroomcnt', 'bathroomcnt','taxamount','calculatedfinishedsquarefeet'], axis=1)
Y = df_train['logerror']
df_prop.bedroomcnt=df_prop.bedroomcnt.fillna(df_prop.bedroomcnt.mean())
df_prop.calculatedfinishedsquarefeet=df_prop.calculatedfinishedsquarefeet.fillna(df_prop.calculatedfinishedsquarefeet.mean())
df_prop.garagetotalsqft=df_prop.garagetotalsqft.fillna(df_prop.garagetotalsqft.mean())
df_prop.bathroomcnt=df_prop.bathroomcnt.fillna(df_prop.bathroomcnt.mean())
df_prop.taxamount=df_prop.taxamount.fillna(df_prop.taxamount.mean())
pred_x = df_prop.filter(['bedroomcnt', 'bathroomcnt','taxamount','calculatedfinishedsquarefeet'], axis=1)
regressor = regressor.fit(X,Y)
pred_y = regressor.predict(pred_x)
score_decision_tree=regressor.score(X,Y)
print("Score : ",score_decision_tree)

pred_x_test = df_test.filter(['bedroomcnt', 'bathroomcnt','taxamount','calculatedfinishedsquarefeet'], axis=1)
pred_y_test = regressor.predict(pred_x_test)
actual = df_test.logerror
#print(actual)

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(actual, pred_y_test)
print("Mean Squared Error :",mse)
```

```python
#K NEAREST NEIGHBOUR MODEL
from sklearn.neighbors import KNeighborsRegressor
x_columns = ['bedroomcnt','bathroomcnt','taxamount']
y_column = ["logerror"]
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(df_train[x_columns], df_train[y_column])
predictions = knn.predict(df_train[x_columns])

actual = df_train[y_column]
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(actual, predictions)
print("Mean Squared Error :",mse)
```

```python
'Which of your models minimizes the squared error?'
Mean Squared Error(MSE) shows the difference between the actual value and the predicted value. It is the average of the squares of the deviations.
Lower MSE score is better. The K nearest neighbour model has a MSE of 0.02 as against the decison tree model which has a MSE of 0.05
So, out of the 2 model, for this data, the K nearest model has a smaller MSE value and is likely to produce better predictions.
```
